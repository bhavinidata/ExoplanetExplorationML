```python
import pandas as pd
import numpy as np
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("Resources/exoplanet_data.csv")
df.head()
```

```python
df.columns
```

```python
df.count()
```

```python
df.dtypes
```

```python
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
df.count()
```

# Select your features (columns)

```python
# Set features. This will also be used as x values. Here all of the columns except "koi_disposition" are selected as features. 
selected_features = df.drop("koi_disposition", axis=1)
selected_features.head()
```

```python
X = selected_features
y = df["koi_disposition"]
```

# Create a Train Test Split

Using `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1, stratify =y)
```

```python
X_train.head()
```

```python
y.head()
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale data using min-max scalar
X_train_scaled = MinMaxScaler().fit(X_train)
X_test_scaled = X_train_scaled.transform(X_test)
X_train_scaled = X_train_scaled.transform(X_train)
```

# Train the Model



### Using SVM without Gridsearch

```python
from sklearn.svm import SVC
svc_model = SVC(kernel='linear')
svc_model.fit(X_train_scaled, y_train)
# predictions = svc_model.predict(X_test)
predictions = svc_model.predict(X_test_scaled)
```

```python
print(f"Training Data Score: {svc_model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {svc_model.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

### Using SVM with Gridsearch
Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
paramgrid = {
                "C": [0.1,1,10,50,100],
                "gamma" : [0.0001, 0.001, 0.01, 0.1]
            }
grid_model = GridSearchCV(svc_model, paramgrid)
```

```python
# Train the model with GridSearch
grid_model.fit(X_train_scaled, y_train)
```

```python
print(grid_model.best_params_)
print(grid_model.best_score_)
```

```python
grid_model.score(X_test_scaled, y_test)
```

```python
predictions = grid_model.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

# Save the Model

```python
# save model with model variable
import joblib
filename = 'bhavini.sav'
joblib.dump(grid_model, filename)
```
